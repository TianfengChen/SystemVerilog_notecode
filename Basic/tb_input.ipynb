```python
import os
import random

a1_n = 'a1.txt'
b1_n = 'a2.txt'
a2_n = 'b1.txt'
b2_n = 'b2.txt'
sel_n= 'sel.txt'
sum_n    = 'sum.txt'
carry_n  = 'carry.txt'

f_a1 = open(a1_n,"w")
f_b1 = open(b1_n,"w")
f_a2 = open(a2_n,"w")
f_b2 = open(b2_n,"w")
f_sel = open(sel_n,"w")
f_sum = open(sum_n,"w")
f_carry = open(carry_n,"w")

for i in range(10):
    a1 = random.randint(0,15)
    b1 = random.randint(0,15)
    a2 = random.randint(0,15)
    b2 = random.randint(0,15)
    sel = random.randint(0,1)
    if sel == 1:
        sum = a2+b2
    else:
        sum = a1+b1
    if sum > 15:
        carry = 1
        sum-=16;
    else:
        carry = 0
    f_a1.write(str(a1))
    f_b1.write(str(b1))
    f_a2.write(str(a2))
    f_b2.write(str(b2))
    f_sel.write(str(sel))
    f_sum.write(str(sum))
    f_carry.write(str(carry))
    if i != 9:
        f_b1.write('\r\n')
        f_a1.write('\r\n')
        f_a2.write('\r\n')
        f_b2.write('\r\n')
        f_sel.write('\r\n')
        f_sum.write('\r\n')
        f_carry.write('\r\n')
```

```python
import os
import random

ipt_name = 'testbench.txt'
opt_name = 'output.txt'
f2  = open(opt_name,"w")
with open(ipt_name,'w') as f:
    for i in range(10):
        a1 = random.randint(0,15)
        b1 = random.randint(0,15)
        a2 = random.randint(0,15)
        b2 = random.randint(0,15)
        sel = random.randint(0,1)
        if sel == 1:
            sum = a2+b2
        else:
            sum = a1+b1
        if sum > 15:
            carry = 1
            sum-=16;
        else:
            carry = 0
        f.write(str(a1)+' '+str(a2)+' '+str(b1)+' '+str(b2)+' '+str(sel)+'\r\n')
        f2.write(str(sum)+' '+str(carry)+'\r\n')

f.close()
f2.close()
```
